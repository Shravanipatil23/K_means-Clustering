```python
import numpy as np
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings("ignore")
```

```python
data_points = np.array([(185,72),(170,56),(168,60),(179,68),(182,72),(180,77),(180,71),(180,70),(183,84),(180,88),(180,67),(177,76)])
```

```python
initial_centers = np.array([(185,72),(170,56)])
```

```python
k = 2
```

```python
kmeans_model = KMeans(n_clusters = k, init = initial_centers, n_init= 10)
```

```python
kmeans_model.fit(data_points)
```

```python
labels = kmeans_model.labels_
labels
```

```python
centroids = kmeans_model.cluster_centers_
centroids
```

```python
from matplotlib import pyplot as plt
```

```python
data_points
```

```python
dp_X = data_points[::,0]
dp_Y = data_points[::,1]
center_X = centroids[::,0]
center_Y = centroids[::,1]
```

```python
plt.scatter(dp_X,dp_Y, c=labels, label = "datapoints")
plt.scatter(center_X,center_Y, c="red", marker = "x" ,label = "centroids")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Heights Vs Weight")
plt.legend()
plt.grid()
```
